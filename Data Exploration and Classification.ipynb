```python
from ucimlrepo import fetch_ucirepo

# Fetch dataset
secondary_mushroom = fetch_ucirepo(id=848)

# Get features and target
X = secondary_mushroom.data.features
y = secondary_mushroom.data.targets
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Encode all categorical features
X_encoded = X.copy()
label_encoders = {}

for col in X_encoded.columns:
    le = LabelEncoder()
    X_encoded[col] = le.fit_transform(X_encoded[col])
    label_encoders[col] = le

# Encode the target column
y_encoded = LabelEncoder().fit_transform(y.values.ravel())

# Combine features and target into one DataFrame
df_pair = X_encoded.copy()
df_pair['target'] = y_encoded

# Use a sample to make it faster for plotting
plot_cols = df_pair.columns[:5].tolist() + ['target']  # first 5 features + target
subset = df_pair[plot_cols].sample(300, random_state=1)

# Enable inline plotting in Jupyter
%matplotlib inline

# Generate the pairplot
sns.pairplot(subset, hue='target')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# Step 1: Encode categorical features
X_encoded = X.copy()
label_encoders = {}

for col in X_encoded.columns:
    le = LabelEncoder()
    X_encoded[col] = le.fit_transform(X_encoded[col])
    label_encoders[col] = le

# Encode the target variable
target_le = LabelEncoder()
y_encoded = target_le.fit_transform(y.values.ravel())

# Step 2: Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

# Step 3: Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# Check shapes
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

# Confusion matrix and report
cm_knn = confusion_matrix(y_test, y_pred_knn)
print(classification_report(y_test, y_pred_knn))
```

```python
sns.heatmap(cm_knn, annot=True, fmt='d', cmap='Greens', xticklabels=['Edible', 'Poisonous'], yticklabels=['Edible', 'Poisonous'])
plt.title("KNN Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
accuracies = []
for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    score = knn.score(X_test, y_test)
    accuracies.append(score)

plt.plot(range(1, 21), accuracies)
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. k in KNN')
plt.grid(True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Train default Decision Tree
dt_default = DecisionTreeClassifier(random_state=42)
dt_default.fit(X_train, y_train)
y_pred_dt_default = dt_default.predict(X_test)

# Evaluate default model
cm_default = confusion_matrix(y_test, y_pred_dt_default)
print("Default Decision Tree Report:")
print(classification_report(y_test, y_pred_dt_default))

# Tuned Decision Tree with max_depth and min_samples_split
dt_tuned = DecisionTreeClassifier(max_depth=5, min_samples_split=10, random_state=42)
dt_tuned.fit(X_train, y_train)
y_pred_dt_tuned = dt_tuned.predict(X_test)

# Evaluate tuned model
cm_tuned = confusion_matrix(y_test, y_pred_dt_tuned)
print("Tuned Decision Tree Report:")
print(classification_report(y_test, y_pred_dt_tuned))

# Plot Confusion Matrices
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
sns.heatmap(cm_default, annot=True, fmt='d', cmap='Blues', ax=ax[0])
ax[0].set_title("Default Decision Tree")
ax[0].set_xlabel("Predicted")
ax[0].set_ylabel("Actual")

sns.heatmap(cm_tuned, annot=True, fmt='d', cmap='Greens', ax=ax[1])
ax[1].set_title("Tuned Decision Tree")
ax[1].set_xlabel("Predicted")
ax[1].set_ylabel("Actual")

plt.tight_layout()
plt.show()

# Plot tuned tree
plt.figure(figsize=(16, 6))
plot_tree(dt_tuned, filled=True, feature_names=X.columns, class_names=['Edible', 'Poisonous'])
plt.title("Optimized Decision Tree")
plt.show()

# Feature importance
importances = dt_tuned.feature_importances_
for feature, importance in zip(X.columns, importances):
    print(f"{feature}: {importance:.3f}")
```
